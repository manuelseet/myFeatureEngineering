```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

import pca_plot
```

Using the wine dataset

```python
df = pd.read_csv("wine.csv")
print(df.shape)
df
```

Plotting the distribution curves of 4 selected features/attributes

```python
sb.set(style='darkgrid')
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

sb.kdeplot(data=df, x='Alcalinity of ash', hue='Cultivar', ax=ax[0,0])
sb.kdeplot(data=df, x='Flavanoids', hue='Cultivar', ax=ax[0,1])
sb.kdeplot(data=df, x='Hue', hue='Cultivar', ax=ax[1,0])
sb.kdeplot(data=df, x='Proline', hue='Cultivar', ax=ax[1,1])
```

```python
X = np.array(df.drop("Cultivar", axis = 1))
scaler = StandardScaler()

Xscaled = scaler.fit_transform(X)
Xscaled.shape
```

Examine the pca plot showing the explained variances of principal components and the cumulative explained variance

```python
pca_plot.pca_plot(Xscaled)
```

Test that the ThresholdValueError exception is thrown with an wrong value for threshold

```python
pca_plot.pca_plot(Xscaled, n_comp=10, threshold=11)
```

Now test for 10 components and with a higher threshold of .90

```python
pca_plot.pca_plot(Xscaled, n_comp=10, threshold=0.9)
```
